```python
import numpy as np
```

```python
df = pd.DataFrame({'관측값_A':[1,-2,5,3,-1,7],
                   '관측값_B':[3,4,np.NaN,9,2,5]})
df
```

# 3-1. df 의 각 행마다 누적합을 반환하세요.

```python
df.cumsum()
```

# 3-2. df 에 관측값_A,B_누적값 열을 추가하세요.(2가지 방법으로)

첫번째 방법

```python
df['관측값_A_누적값'] = df['관측값_A'].cumsum()
df
```

```python
df['관측값_B_누적값'] = df['관측값_B'].cumsum()
df
```

```python
df[['관측값_A','관측값_A_누적값','관측값_B','관측값_B_누적값']]
df
```

두번째 방법

```python
df2 = pd.DataFrame({'관측값_A':[1,-2,5,3,-1,7],
                   '관측값_B':[3,4,np.NaN,9,2,5]})
df2
```

```python
df2.insert(df2.columns.get_loc('관측값_A')+1,
          '관측값_A_누적값',df2['관측값_A'].cumsum())
df2
```

```python
df2.insert(df2.columns.get_loc('관측값_B')+1,
          '관측값_B_누적값',df2['관측값_B'].cumsum())
df2
```

# 3-3. df 에 관측값_A,B_누적곱 열을 추가하세요.

```python
df.insert(df.columns.get_loc('관측값_A_누적값')+1,
         '관측값_A_누적곱',df['관측값_A'].cumprod())
df
```

```python
df.insert(df.columns.get_loc('관측값_B_누적값')+1,
         '관측값_B_누적곱',df['관측값_B'].cumprod())
df
```
