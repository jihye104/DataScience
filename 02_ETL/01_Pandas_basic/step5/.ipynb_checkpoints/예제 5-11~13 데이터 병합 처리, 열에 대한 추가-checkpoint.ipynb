## 예제 5-11  데이터를 병합 처리하기

```python
data = {

"student_id": [11111, 11112, 11113, 11114],

"Math": [60, 80, 90, 75],

"English": [85, 90, 65, 70],

"Class": [1, 2, 3, 4]

}
```

```python
df = pd.DataFrame(data)
df
```

melt : 병합
id_vars : 고정열,  
var_name : 병합한 열들의 새로운 열이름,  
value_name : 병합할 열의 새로운 값 이름,  
value_vars : 병합할 Target 열

```python
df2 = df.melt(id_vars = 'student_id',
             var_name = 'Subject',
             value_name = 'Score',
             value_vars = ['Math','English'])
df2
```

```python
df3 = df.melt(id_vars = ["student_id","Class"],
              var_name = "Subject",
              value_name = "Score",
              value_vars=['Math', 'English'])
df3
```

## 예제 5-12 열에 대한 추가

```python
import numpy as np
```

```python
df = pd.DataFrame({'A': range(1, 11),
                   'B': np.random.randn(10)})
df
```

```python
df_1 = df.copy()
```

assign => 기존의 열이 없으면 지정한 이름으로 갱신  
lambda 메개변수 : 반환하는 값

```python
df_1.assign(In_A = lambda x: np.log(x.A))
```

```python
df_2 = df.copy()
```

기존에 했던 방식

```python
df_2['In_A'] = np.log(df_2['A'])
df2
```

## 예제 5-13 열에 대한 추가 활용 

```python
movie = pd.read_csv('../data/korea_movie_list.csv',encoding='cp949')
movie.head()
```

```python
movie_ = pd.concat([movie.filter(like = 'prod'),
                   movie.filter(like = 'title')],
                  axis = 1)
movie_.head()
```
