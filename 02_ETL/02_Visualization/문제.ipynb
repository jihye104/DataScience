### 문제3.

```python
x = []
y = []

with open("test.txt", "r") as file : 
    lines = file.readlines()
    
    for line in lines :
        line = line.strip()
        x.append(int(line.split(' ')[0]))
        y.append(int(line.split(' ')[1]))
print(x)
print(y)

plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.margins(x=0, y=0)

plt.title('Sample graph!')

plt.show()
```

### 문제5.

```python
x1 = [10, 20, 30]
y1 = [20, 40, 10]

x2 = [10, 20, 30]
y2 = [40, 10, 30]

plt.xlabel('x - axis')
plt.ylabel('y - axis')

plt.title('Two or more lines on same plot with suitable legends')

plt.margins(x=0, y=0)

plt.plot(x1,y1,'dodgerblue', label='line 1')
plt.plot(x2,y2,'limegreen', label='line 2')

plt.legend()
plt.show()
```

### 문제6.

```python
x1 = [10, 20, 30]
y1 = [20, 40, 10]

x2 = [10, 20, 30]
y2 = [40, 10, 30]

plt.xlabel('x - axis')
plt.ylabel('y - axis')

plt.title('Two or more lines on same plot with suitable legends')

plt.margins(x=0, y=0)

plt.plot(x1,y1,'blue', linewidth = 3, label='line 1-width-3')
plt.plot(x2,y2,'red', linewidth = 5, label='line 2-width-5')

plt.legend()
plt.show()
```

### 문제7.

```python
x1 = [10, 20, 30]
y1 = [20, 40, 10]

x2 = [10, 20, 30]
y2 = [40, 10, 30]

plt.xlabel('x - axis')
plt.ylabel('y - axis')

plt.title('Two or more lines on same plot with suitable legends')

plt.margins(x=0, y=0)

plt.plot(x1,y1,'blue', linewidth = 3, label='line 1-dotted', linestyle='dotted')
plt.plot(x2,y2,'red', linewidth = 5, label='line 2-dashed', linestyle='dashed')

plt.legend()
plt.show()
```

### 문제 8.

```python
x = [1,4,5,6,7]
y = [2,6,3,6,3]

plt.plot(x, y, color='red', linestyle='dashdot', marker='o', markerfacecolor='blue', markersize=14)

plt.ylim(1,8)
plt.xlim(1,8)

plt.xlabel('x - axis')
plt.ylabel('y - axis')

plt.title('Display marker')
plt.show()
```
