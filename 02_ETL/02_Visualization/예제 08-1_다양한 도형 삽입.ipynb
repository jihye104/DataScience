# 다양한 도형 삽입하기

#### matplotlib.patches 모듈은 그래프에 다양한 2D 도형을 표현하기 위한 클래스를 포함

### 사각형 그리기 (Rectangle)

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
```

patches.Rectangle((x,y),width,height)  
사각형 그리기 할때는 (x,y) 이렇게 괄호 해줘야함

```python
# 1. 기본 스타일 설정
plt.style.use('default')
plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['font.size'] = 12

# 2. 데이터 준비
np.random.seed(0)
data_a = np.random.normal(0, 2.0, 1000)
data_b = np.random.normal(-3.0, 1.5, 500)
data_c = np.random.normal(1.2, 1.5, 1500)

# 3. 그래프 그리기
fig, ax = plt.subplots()

violin = ax.violinplot([data_a, data_b, data_c], showmeans=True)
ax.set_ylim(-10.0, 10.0)
ax.set_xticks(np.arange(1, 4))
ax.set_xticklabels(['A', 'B', 'C'])
ax.set_xlabel('Data Type')
ax.set_ylabel('Distribution')

violin['bodies'][0].set_facecolor('blue')
violin['bodies'][1].set_facecolor('red')
violin['bodies'][2].set_facecolor('green')

violin['cbars'].set_edgecolor('gray')
violin['cmaxes'].set_edgecolor('gray')
violin['cmins'].set_edgecolor('gray')
violin['cmeans'].set_edgecolor('gray')

# 4. 사각형 그리기
ax.add_patch(
    patches.Rectangle(
        (1.8, 1.0), 0.4, 1.5,
        edgecolor = 'deeppink',
        facecolor = 'lightgray',
        fill=True
    )
)

# 5. 텍스트 삽입하기
ax.text(1.5, 3.5, 'Max of Data B')

plt.show()
```

### 화살표 그리기 (Arrow)

patches.Arrow(x,y,dx(현재 설정된 x 값을 기준으로 0),dy(화살표 길이(양수면 ↑, 음수면 ↓)), width = 화살표 넓이)

```python
# 1. 기본 스타일 설정
plt.style.use('default')
plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['font.size'] = 12

# 2. 데이터 준비
np.random.seed(0)
data_a = np.random.normal(0, 2.0, 1000)
data_b = np.random.normal(-3.0, 1.5, 500)
data_c = np.random.normal(1.2, 1.5, 1500)

# 3. 그래프 그리기
fig, ax = plt.subplots()

violin = ax.violinplot([data_a, data_b, data_c], showmeans=True)
ax.set_ylim(-10.0, 10.0)
ax.set_xticks(np.arange(1, 4))
ax.set_xticklabels(['A', 'B', 'C'])
ax.set_xlabel('Data Type')
ax.set_ylabel('Value')

violin['bodies'][0].set_facecolor('blue')
violin['bodies'][1].set_facecolor('red')
violin['bodies'][2].set_facecolor('green')

violin['cbars'].set_edgecolor('gray')
violin['cmaxes'].set_edgecolor('gray')
violin['cmins'].set_edgecolor('gray')
violin['cmeans'].set_edgecolor('gray')

# 4. 화살표 그리기
ax.add_patch(
    patches.Arrow(
        2,4,0,-1.5,
        width = 0.5,
        edgecolor = 'deeppink',
        facecolor = 'lightgray'
    )
)

# 5. 텍스트 삽입하기
ax.text(1.5, 5.0, 'Max of Data B')

plt.show()
```

### 다각형 그리기 (Polygon)

patches.Polygon(((x1,y1),(x2,y2),(x3,y3)), closed = True)  
원하는 다각형 꼭지점을 좌표로 지정  
closed = True => 디폴트, False 하면 마지막 선 안그어짐

```python
# 1. 기본 스타일 설정
plt.style.use('default')
plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['font.size'] = 12

# 2. 데이터 준비
np.random.seed(0)
data_a = np.random.normal(0, 2.0, 1000)
data_b = np.random.normal(-3.0, 1.5, 500)
data_c = np.random.normal(1.2, 1.5, 1500)

# 3. 그래프 그리기
fig, ax = plt.subplots()

violin = ax.violinplot([data_a, data_b, data_c], showmeans=True)
ax.set_ylim(-10.0, 10.0)
ax.set_xticks(np.arange(1, 4))
ax.set_xticklabels(['A', 'B', 'C'])
ax.set_xlabel('Data Type')
ax.set_ylabel('Value')

violin['bodies'][0].set_facecolor('blue')
violin['bodies'][1].set_facecolor('red')
violin['bodies'][2].set_facecolor('green')

violin['cbars'].set_edgecolor('gray')
violin['cmaxes'].set_edgecolor('gray')
violin['cmins'].set_edgecolor('gray')
violin['cmeans'].set_edgecolor('gray')

# 4. 다각형 그리기
ax.add_patch(
    patches.Polygon(
        ((1.9, 4),(2, 2.5),(2.1, 4)),
        closed=True,
        edgecolor = 'deeppink',
        facecolor = 'lightgray'
    )
)

# 5. 텍스트 삽입하기
ax.text(1.5, 5.0, 'Max of Data B')

plt.show()
```

### 다양한 패턴 채우기

bar() 함수의 hatch 는 막대 그래프의 막대 내부에 표시할 패턴을 지정  
예제에서는 네가지의 서로 다른 패턴을 지정  
아래의 hatch 문자열 중 선택해서 사용가능  

'/', '\\', '|', '-', '+', 'x', 'o', 'O', '.', '*'

### 예제1

```python
plt.style.use('default')
plt.rcParams['figure.figsize'] = (6, 5)
plt.rcParams['font.size'] = 12

x = [1,2,3]
y = [1,2,3]

fig, ((ax1, ax2),(ax3, ax4)) = plt.subplots(2,2)

ax1.bar(x, y, color='aquamarine', edgecolor='black', hatch='/')
ax2.bar(x, y, color='salmon', edgecolor='black', hatch='\\')
ax3.bar(x, y, color='navajowhite', edgecolor='black', hatch='+')
ax4.bar(x, y, color='lightskyblue', edgecolor='black', hatch='*')

plt.tight_layout()
plt.show()
```

```python
plt.style.use('default')
plt.rcParams['figure.figsize'] = (6, 5)
plt.rcParams['font.size'] = 12

x = [1,2,3]
y = [1,2,3]

fig, ((ax1, ax2),(ax3, ax4)) = plt.subplots(2,2)

ax1.bar(x, y, color='aquamarine', edgecolor='black', hatch='|')
ax2.bar(x, y, color='salmon', edgecolor='black', hatch='-')
ax3.bar(x, y, color='navajowhite', edgecolor='black', hatch='x')
ax4.bar(x, y, color='lightskyblue', edgecolor='black', hatch='o')

plt.tight_layout()
plt.show()
```

```python
plt.style.use('default')
plt.rcParams['figure.figsize'] = (6, 5)
plt.rcParams['font.size'] = 12

x = [1,2,3]
y = [1,2,3]

fig, ((ax1, ax2),(ax3, ax4)) = plt.subplots(2,2)

ax1.bar(x, y, color='aquamarine', edgecolor='black', hatch='O')
ax2.bar(x, y, color='salmon', edgecolor='black', hatch='.')
ax3.bar(x, y, color='navajowhite', edgecolor='black', hatch='\\')
ax4.bar(x, y, color='lightskyblue', edgecolor='black', hatch='*')

plt.tight_layout()
plt.show()
```

### 예제2

```python
plt.style.use('default')
plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['font.size'] = 12

x = [1, 2, 3]
y = [1, 2, 3]

fig, ax = plt.subplots()
bars = ax.bar(x,y, color='lightgray', edgecolor='black')

bars[0].set_hatch('x')
bars[1].set_hatch('O')
bars[2].set_hatch('.')

plt.tight_layout()
plt.show()
```

### 패턴의 밀도 지정하기
hatch 문자열의 개수를 한 개 이상 입력함으로써 패턴의 밀도를 지정

```python
plt.style.use('default')
plt.rcParams['figure.figsize'] = (8, 5)
plt.rcParams['font.size'] = 12

x = [1,2,3]
y = [1,2,3]

for i in range(6):
    hatch_str = '/'*i
    
    ax = plt.subplot(2,3, i+1)
    ax.set_title("Hatch String: " + hatch_str)
    bars = ax.bar(x,y, facecolor = 'skyblue', edgecolor = 'black')
    
    for bar in bars:
        bar.set_hatch(hatch_str)
        
plt.tight_layout()
plt.show()
```
