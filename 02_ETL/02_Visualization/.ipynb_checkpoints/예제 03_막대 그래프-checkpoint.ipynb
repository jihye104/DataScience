## 기본 사용

```python
import matplotlib.pyplot as plt
import numpy as np
```

1. 보여줄 bar의 인덱스 생성  
ex) x = np.arange(3)  
years = ['2018', '2019', '2020']
2. bar에 매칭이 되는 값 리스트 생성  
ex) values = [100, 400, 900]

2018년 누적된 레코드 결과가 100건이라는 의미  
실제 데이터는 2018년 관련 레코드가 100건이 있다는 의미  
2019,2020년도는 각각 400건, 900건이 있다는 의미

```python
x = np.arange(3)
years = ['2018', '2019', '2020']
values = [100, 400, 900]

plt.xticks(x, years)
plt.bar(x, values)

plt.show()
```

### 색상 지정하기

```python
x = np.arange(3)
years = ['2018', '2019', '2020']
values = [100, 400, 900]

plt.xticks(x, years, color='y')
# plt.bar(x, values, color='r')
# plt.bar(x, values, color='dodgerblue')
plt.bar(x, values, color='C2')
# plt.bar(x, values, color='#e35f62')

plt.show()
```

### 바마다 개별 색 지정

색상이름을 리스트로 지정하면 순서대로 바에 할당

```python
x = np.arange(3)
years = ['2018', '2019', '2020']
values = [100, 400, 900]
colors = ['y', 'dodgerblue', 'C2']

plt.bar(x, values, color=colors)
plt.xticks(x, years)

plt.show()
```

### 막대 폭 지정하기

```python
x = np.arange(3)
years = ['2018', '2019', '2020']
values = [100, 400, 900]

plt.xticks(x, years)
# plt.bar(x, values, width=0.4)
# plt.bar(x, values, width=0.6)
# plt.bar(x, values, width=0.8)
plt.bar(x, values, width=1.0)

plt.show()
```

### 스타일 꾸미기

align은 눈금과 막대의 위치를 조절.  
디폴트 값은 center 이며, edge 로 설정하면 막대의 왼쪽 끝에 눈금이 표시  
edgecolor는 막대 테두리 색, linewidth는 테두리의 두께를 지정.  
tick_label을 리스트 또는 어레이 형태로 지정하면, 틱에 문자열을 순서대로 나타냄.

```python
from matplotlib import font_manager,rc
```

```python
font_location = "C:\Windows\Fonts\malgun.ttf"
font_name = font_manager.FontProperties(fname=font_location).get_name()
rc('font',family=font_name)
```

```python
x = np.arange(3)
years = ['2018년', '2019년', '2020년']
values = [100, 400, 900]

plt.bar(x, values, align='edge', edgecolor='lightgray',
        linewidth=5, tick_label=years)

plt.show()
```

## 수평 막대 그래프 그리기

### 기본 사용 (barh 함수 활용)

```python
y = np.arange(3)
years = ['2018', '2019', '2020']
values = [100, 400, 900]

plt.barh(y, values)
plt.yticks(y, years)

plt.show()
```

### 색상지정하기

```python
y = np.arange(3)
years = ['2018', '2019', '2020']
values = [100, 400, 900]

plt.barh(y, values, color='y')
# plt.barh(y, values, color='dodgerblue')
# plt.barh(y, values, color='C2')
# plt.barh(y, values, color='#e35f62')
plt.yticks(y, years)

plt.show()
```

### 색상 개별 지정

```python
y = np.arange(3)
years = ['2018', '2019', '2020']
values = [100, 400, 900]
colors = ['y', 'dodgerblue', 'C2']

plt.barh(y, values, color=colors)
plt.yticks(y, years)

plt.show()
```

### 막대 위치 지정

```python
y = np.arange(3)
years = ['2018', '2019', '2020']
values = [100, 400, 900]

plt.barh(y, values, height=0.4)
# plt.barh(y, values, height=0.6)
# plt.barh(y, values, height=0.8)
# plt.barh(y, values, height=1.0)
plt.yticks(y, years)

plt.show()
```

### 스타일 꾸미기

```python
y = np.arange(3)
years = ['2018', '2019', '2020']
values = [100, 400, 900]

plt.barh(y, values, align='edge', edgecolor='#eee',
         linewidth=5, tick_label=years)

plt.show()
```
