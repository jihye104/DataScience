### 넘파이의 random 함수

### 1. numpy.random.rand
 0~1 사이의 값의 uniform distribution 균일 분포, 균등 분포라고 불리는 값을 return 하는 함수

```python
DATA_LIMIT = 1000000
```

```python
np.random.rand(5)
```

```python
np.random.rand(3,5)
```

```python
rand_data = np.random.rand(DATA_LIMIT)
plt.hist(rand_data)
```

### 2. numpy.random.randn

Normal(Gaussian) Distribution 정규분포, 가우시안 분포라고 불리는 값을 return 하는 함수  
평균 0, 분산이 1인 값을 반환

```python
randn_data = np.random.randn(DATA_LIMIT)
plt.hist(randn_data, bins=100)
```

## 3. numpy.random.randint

[이산균등분포(discrete uniform)](https://ko.wikipedia.org/wiki/%EC%9D%B4%EC%82%B0%EA%B7%A0%EB%93%B1%EB%B6%84%ED%8F%AC)

1) low : 가장 작은 범주가 어디서부터 시작하는지를 적는다.  
단, 1개의 hyperparameter 를 적으면 그 값은 가장 높은 범주의 값으로 인식하고 0에서부터 그 숫자 사이의 int 를 1개 반환한다.  
만약 high 없이 개수를 늘리고 싶다면 다음과 같이 적으면 된다.  
numpy.random.randint(3, size=10)  
  
2) high : 가장 높은 정수를 적은 곳으로 선택사항 optional 입니다.  
  
3) size : 몇 개의 random 값을 리턴하기 바라는지를 적으면 됨  
3개 반환을 받고 싶다면 3개를 적으면 된다.  
만약 array 형태로 shape 를 적고싶다면 다음과 같이 적을 수도 있다.  
numpy.random.randint(3,10, size = (3,2))  
이렇게 적으면 3~10 사이의 정수를 3*2 행렬에 맞게 반환하라는 뜻  
  
4) dtype : int 를 반환하는 함수이기 때문에 굳이 필요한가 싶지만, 이부분은 int 혹은 int64 를 반환할 수 있또록 선택하는 parameter 입니다.  
선택하지 않아도 무방

### high만 지정

```python
np.random.randint(3,size = 10)
```

### low, high 지정

```python
np.random.randint(1,3,size = 10)
```

```python
randint_data = np.random.randint(1,1000, size = DATA_LIMIT)
plt.hist(randint_data, bins = 20)
```

```python
move_right =  173
variance_amplication = 5

height_data = np.round(
    move_right+
    (np.random.randn(DATA_LIMIT)*variance_amplication),
    1
)

height_data
```

```python
plt.xlabel('키')
plt.ylabel('누적개수')
plt.hist(height_data, bins=40)
```

```python
m_move_right =  173
w_move_right =  161
variance_amplication = 5
m_height_data = np.round(m_move_right + (np.random.randn(DATA_LIMIT)*variance_amplication),1)
w_height_data = np.round(w_move_right + (np.random.randn(DATA_LIMIT)*variance_amplication),1)

plt.xlabel('키')
plt.ylabel('누적개수')
plt.hist((m_height_data,w_height_data), bins=100, label=['성인 남자키','성인 여자키'])

plt.title('전국 성인남녀 키 현황 (2022.3 산업통산자원부 통계자료)')
plt.legend()

plt.show()
```
