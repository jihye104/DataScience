```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import font_manager,rc

import random

font_location = "C:\Windows\Fonts\malgun.ttf"
font_name = font_manager.FontProperties(fname=font_location).get_name()
rc('font',family=font_name)
plt.rcParams['axes.unicode_minus'] = False
```

# 1.
x 축, y 축 및 제목에 적절한 레이블이 있는 선을 그리는 Python 프로그램을 작성하십시오.

```python
plt.rcParams['xtick.top'] = True
plt.rcParams['ytick.right'] = True
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'

x = np.arange(1,50)
# y = [value*3 for value in x]
y = x*3

plt.plot(x,y)

plt.title('Draw a line')
plt.xlabel('x - axis')
plt.ylabel('y - axis')

plt.xlim(0,50)
plt.ylim(0,160)

plt.show()
```

# 2.
x 축, y 축 및 제목에 적절한 레이블이 있는 지정된 축 값을 사용하여 선을 그리는 Python 프로그램을 작성하십시오.

```python
plt.rcParams['xtick.top'] = True
plt.rcParams['ytick.right'] = True
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'

x = [1,2,3]
y = [2,4,1]

plt.plot(x,y)

plt.title('Sample graph!')
plt.xlabel('x - axis')
plt.ylabel('y - axis')

plt.xlim(1.0,3.0)
plt.ylim(1.0,4.0)

plt.show()
```

# 3.
2016년 10월 3일부터 2016년 10월 7일까지 Alphabet Inc.의 재무 데이터에 대한 선형 차트를 그리는 Python 프로그램을 작성하십시오.

sep => 구분자  
parse_dates = True => 데이터를 읽으면서 날짜를 DatetimeIndex 객체로 변경하는 유용한 옵션  
index_col = 0 => Unnamed: 0 컬럼이 없어짐

```python
import matplotlib.dates as mdates

plt.rcParams['xtick.top'] = False
plt.rcParams['ytick.right'] = False
plt.rcParams['xtick.direction'] = 'out'
plt.rcParams['ytick.direction'] = 'out'

df = pd.read_csv('fdata.csv', sep = ',', parse_dates = True, index_col = 0)

df.plot()

plt.xlabel(' Oct\n2016', loc = 'left')

plt.show()
```

# 4.
각 라인의 적절한 범례와 함께 동일한 플롯에 두 개 이상의 라인을 플로팅하는 Python 프로그램을 작성하십시오.

```python
plt.rcParams['xtick.top'] = True
plt.rcParams['ytick.right'] = True
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'

x1 = [10,20,30]
y1 = [20,40,10]
y2 = [40,10,30]

plt.plot(x1,y1, 'b-', label = 'line 1')
plt.plot(x1,y2, 'g-', label = 'line 2')

plt.title('Two or more lines on same plot with suitable legends')
plt.xlabel('x - axis')
plt.ylabel('y - axis')

plt.xlim(10,30)
plt.ylim(10,40)

plt.legend(loc = 'best', fontsize = 12, frameon = True, edgecolor = 'black')
plt.show()
```

# 5.
범례, 너비 및 색상이 다른 두 개 이상의 선을 플로팅하는 Python 프로그램을 작성하십시오.

```python
plt.rcParams['xtick.top'] = True
plt.rcParams['ytick.right'] = True
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'

x1 = [10,20,30]
y1 = [20,40,10]
y2 = [40,10,30]

plt.plot(x1,y1, 'b-', label = 'line1-width-3', linewidth = 3)
plt.plot(x1,y2, 'r-', label = 'line2-width-5', linewidth = 5)

plt.title('Two or more lines with different widths and colors with suitable legends')
plt.xlabel('x - axis')
plt.ylabel('y - axis')

plt.xlim(10,30)
plt.ylim(10,40)

plt.legend(loc = 'best', fontsize = 12, frameon = True, edgecolor = 'black')
plt.show()
```

# 6.
스타일이 다른 두 개 이상의 선을 플로팅하는 Python 프로그램을 작성합니다.

```python
plt.rcParams['xtick.top'] = True
plt.rcParams['ytick.right'] = True
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'

x1 = [10,20,30]
y1 = [20,40,10]
y2 = [40,10,30]

plt.plot(x1,y1, label = 'line1-dotted', linestyle = 'dotted', linewidth = 2, color = 'blue')
plt.plot(x1,y2, label = 'line2-dashed', linestyle = 'dashed', linewidth = 3, color = 'red')

plt.title('Plot with two or more lines with different styles')
plt.xlabel('x - axis')
plt.ylabel('y - axis')

plt.xlim(10,30)
plt.ylim(10,40)

plt.legend(loc = 'best', fontsize = 12, frameon = True, edgecolor = 'black')
plt.show()
```

# 7.
두 개 이상의 라인을 플로팅하고 라인 마커를 설정하는 Python 프로그램을 작성하십시오

```python
x = [1,4,5,6,7]
y = [2,6,3,6,3]

plt.plot(x,y, 'r-.', linewidth = 3, marker = 'bo', markersize = 12)

plt.xlim(1,8)
plt.ylim(1,8)
```
