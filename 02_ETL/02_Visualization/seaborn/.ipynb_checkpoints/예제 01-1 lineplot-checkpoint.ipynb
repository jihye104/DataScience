```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt## Seaborn 장점
- 통계관점의 보조지표를 자동으로 제공한다.
- 그룹별 차트 생성이 용이하다
```

환자별 subject 시가점마다의 fmri signal 값을 담고 있는 데이터

```python
fmri = sns.load_dataset('fmri')
fmri
```

```python
sns.set(style = 'darkgrid')
sns.lineplot(x = 'timepoint', y = 'signal', data = fmri)
```

* 중앙에 있는 선은 각 timepoint 별 signal 의 평균값을 표현하고 있고, 넓은 영역은 각 timepoint 별 signal 의 분포(범위)를 표현  
* 대략 3.5초까지는 환자별로 거의 동이란 fmri signal 을 보여주지만, 이후 부터 7.5초까지는 꽤 큰 분포가 있는 것을 확인

```python
fmri['event'].value_counts()
```

hue 를 통해 분류별 차트를 자동으로 생성할 수 있다.

```python
sns.lineplot(x = 'timepoint', y = 'signal', hue = 'event', data = fmri)
```
