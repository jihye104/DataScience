```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import font_manager,rc
```

```python
font_location = "C:\Windows\Fonts\malgun.ttf"
font_name = font_manager.FontProperties(fname=font_location).get_name()
rc('font',family=font_name)
```

```python
month = np.arange(1,13)
b_weight = [1,1.5,1.8,2,2.3,2.3,2.5,2.8,2.7,3,2.8,3.2]
w_weight = [1,1.8,2,2.2,2.5,2.8,3.5,3.8,4,4.1,4.3,4.5]
plt.plot(month,b_weight,label = '뽀짝', color='#ff99ff',
         marker='o', linestyle='-')
plt.plot(month,w_weight,label = '겨울', color='#ff6600',
         marker='*', linestyle='-')

plt.xlabel('월',labelpad=15)
plt.ylabel('몸무게',labelpad=15)

plt.legend(loc = 'best', ncol = 2, fontsize = 10, shadow=True)

plt.show()
```
